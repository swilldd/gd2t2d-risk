### Decoding categorical features

We will not  decode the target variable or `socioeconomic_factors_deprivation_quintile` as this was how the data was referenced in the paper. We don't vhave meanings for those values.

These values will be encoded again during model building.

```python
df_decoded = df.copy()
```

```python
df_decoded.columns
```

```python

# Binary features: 0 = no, 1 = yes
binary_features = [
    'older_maternal_age', 'high_pre_pregnancy_bmi_or_overweight',
    'family_history_of_diabetes', 'presence_of_t2dm_associated_gene_variants',
    'multiparity', 'insulin_treatment_during_pregnancy',
    'pregnancy_complications_hypertensive_disorders',
    'pregnancy_complications_preterm_delivery',
    'pregnancy_complications_pph', 'gestational_weight_gain',
    'large_for_gestational_age',
    'macrosomia_baby_birth_weightdelivered_a_baby_greater_than_3_5kg',
    'instrumental_delivery', 'stillbirth_miscarriage', 'nicu_admission',
    'perinatal_outcome_28_day_mortality',
    'obesity_or_unhealthy_postpartum_weight_gain', 'physical_inactivity',
    'unhealthy_diet', 'smoking', 'alcohol_intake',
    'does_not_undergo_postpartum_glucose_screening', 'breastfeeding',
    'history_of_recurrence_of_gdm'
]

binary_map = {0: 'no', 1: 'yes'}

for feature in binary_features:
    df_decoded[feature] = df_decoded[feature].map(binary_map)

# Ethnicity
ethnicity_values = {
    0: 'other',
    1: 'mixed',
    2: 'black',
    3: 'asian',
    4: 'white'
}

df_decoded.ethnicity = df_decoded.ethnicity.map(ethnicity_values)

# Abnormal OGTT results
ogtt_values = {
    0: 'normal',
    1: 'abnormal'
}

df_decoded.abnormal_ogtt_results = df_decoded.abnormal_ogtt_results.map(ogtt_values)

# Elevated HbA1c during pregnancy
hba1c_values = {
    0: 'normal',
    1: 'elevated'
}

df_decoded.elevated_hba1c_during_pregnancy = df_decoded.elevated_hba1c_during_pregnancy.map(hba1c_values)

```

### Feature Importance

As all features are categorical we can use the following methods to measure feature importance:
- mutual information score

This can tell us how strongly a feature and the target are related.

```python
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df.older_maternal_age, df.t2dm_risk)
```

```python
feature_names = list(df.columns)
```

```python
feature_names = feature_names[:-1]
```

```python
feature_names
```

```python
for category in feature_names:
    mi_score = mutual_info_score(df[category], df.t2dm_risk)
    print(f"{category.capitalize()} mutual information score: {mi_score:.3f}")
    print()
```

```python
def mutual_info_risk_score(series: pd.Series):
    return mutual_info_score(series, df.t2dm_risk)
```

```python
df[feature_names].apply(mutual_info_risk_score).sort_values(ascending=False)
```

Whether or not a patient underwent insulin treatment during pregnancy has the most mutual information an so may be one of the most influential features when predicting risk of developing type II diabetes. THis is followed by:
- elevated_hba1c_during_pregnancy
- large_for_gestational_age   
- macrosomia_baby_birth_weightdelivered_a_baby_greater_than_3_5kg  

In contrast, factors like ethnicity, smoking, alcohol intake, perinatal outcomes, and stillbirth/miscarriage contribute virtually no predictive information for T2DM in this dataset.
